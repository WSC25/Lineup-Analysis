```python
data.rename(columns={'得分': 'RUN'}, inplace=True)
```

```python
data['一安'] =data['安打']-data['二壘打']-data['三壘打']-data['全壘打']
data['XB'] =data['二壘打']+data['三壘打']+data['全壘打']
data['長打/PA'] =data['XB']/data['打席']
data['安打/PA'] = data['安打']/data['打席']
data['OPS'] =data['上壘率'] + data['長打率']
data['BB%'] = (data['四壞球']/data['打席'])*100
data['K%'] = (data['三振']/data['打席'])*100
data['TB/PA'] = data['壘打數']/data['打席']
data
```

```python
#棒次數據整理
agg_dict = {
    "RUN": "sum",
    }
score_total = data.groupby(["場次",'日期','球隊']).agg(agg_dict).reset_index()
score_total
```

```python
df = score_total.merge(data, on=["場次",'日期','球隊'])
df.rename(columns={'RUN_x': '得分'}, inplace=True)
df
```

```python
# 轉換棒次為字串並去掉 .0
df = df[df["棒次"] != 0.0]
#df.dropna()
df["棒次"] = df["棒次"].astype(str).str.replace(r"\.0$", "", regex=True)
```

```python
excel_file_path5 = 'C:/Users/ASUS/OneDrive/Documents/中信兄弟/CPBL/打序/rebas_stat.xlsx'
sheet_name1 = 'Sheet 1' 
data5 = pd.read_excel(excel_file_path5, sheet_name=sheet_name1)
data5
```

```python
data5['TB/PA'] = data5['壘打數']/data5['打席']
```

```python
col = ["場次",'日期','球隊','得分','棒次','TB/PA','XB',"打擊率", "上壘率", "長打率", "OPS", "BB%", "K%",'長打/PA','安打/PA']
data5 = data5[col]
df=df[col]

merge = pd.concat([data5, df], ignore_index=True)
merge
```

```python
# 先將非數字的 '棒次' 值轉為 NaN（防止有 'nan' 這種字串）
merge["棒次"] = pd.to_numeric(merge["棒次"], errors='coerce')

# 刪除 NaN 行
merge = merge.dropna(subset=["棒次"])

# 轉換為整數
merge["棒次"] = merge["棒次"].astype(int)
```

```python
# 假設 df 是你的原始數據
#df = rebas_stat  # 你的 DataFrame

# 重新排列數據，使不同打序的統計數據變成獨立欄位
df_pivot = merge.pivot_table(index=["場次",'日期','球隊','得分'], 
                     columns="棒次", 
                     values=['TB/PA','XB',"打擊率", "上壘率", "長打率", "OPS", "BB%", "K%",'長打/PA','安打/PA'])

# 重新命名欄位，例如 BA_1, OBP_1, ..., K%_9
df_pivot.columns = [f"{col}{order}" for col, order in df_pivot.columns]
df_pivot = df_pivot.reset_index()

# 顯示結果
df_pivot = df_pivot.fillna(0)
df_pivot
```

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 假設 df_pivot 是已經整理好的數據
#df = df_pivot.copy()

# 定義 X（所有打序的 BA、OBP、SLG、OPS、BB%、K%）和 Y（得分）
X_cols = [col for col in  df_pivot.columns if any(stat in col for stat in ['TB/PA','上壘率', "OPS", "BB%", "K%",'長打/PA','安打/PA'])]
X =  df_pivot[X_cols]
Y =  df_pivot["得分"]

# 添加常數項（截距）
X = sm.add_constant(X)

# 執行回歸分析
model = sm.OLS(Y, X).fit()
print(model.summary())

```

```python
# 計算 VIF（方差膨脹因子）檢測多重共線性
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# 顯示 VIF 結果
vif_data
```

```python
# 假設你已經有回歸模型 model 並提取其係數
import itertools
params = model.params
coef = {
    "const": params["const"],
    "BB%": [params[f"BB%{i}"] for i in range(1, 10)],
    "上壘率": [params[f"上壘率{i}"] for i in range(1, 10)],
    "TB/PA": [params[f"TB/PA{i}"] for i in range(1, 10)],
    "OPS": [params[f"OPS{i}"] for i in range(1, 10)],
    "K%": [params[f"K%{i}"] for i in range(1, 10)],
    "長打/PA": [params[f"長打/PA{i}"] for i in range(1, 10)],
    "安打/PA": [params[f"安打/PA{i}"] for i in range(1, 10)]
}

# 讀取打者數據
# 讀取打者數據
batters = [
    {"name": "王博玄", "AVG":0.283 ,"OBP": 0.332, "SLG": 0.355, "K%": 0.171,"OPS":0.686},
    {"name": "陳文杰", "AVG":0.265 ,"OBP": 0.332, "SLG": 0.376, "K%": 0.145,"OPS":0.708},
    {"name": "陳傑憲", "AVG":0.334 ,"OBP": 0.421, "SLG": 0.416, "K%": 0.106,"OPS":0.837},
    {"name": "吳念庭", "AVG":0.240 ,"OBP": 0.295, "SLG": 0.310, "K%": 0.165,"OPS":0.605},
    {"name": "高宇杰", "AVG":0.252 ,"OBP": 0.316, "SLG": 0.330, "K%": 0.167,"OPS":0.646},
    {"name": "張育成", "AVG":0.270 ,"OBP": 0.445, "SLG": 0.540, "K%": 0.220,"OPS":0.985},
    {"name": "蔣少宏", "AVG":0.247 ,"OBP": 0.306, "SLG": 0.348, "K%": 0.139,"OPS":0.654},
    {"name": "宋晨睿", "AVG":0.341 ,"OBP": 0.396, "SLG": 0.444, "K%": 0.220,"OPS":0.840},
    {"name": "林子偉", "AVG":0.193 ,"OBP": 0.269, "SLG": 0.319, "K%": 0.196,"OPS":0.588}
]

# 計算所有可能的打線排列
lineup_scores = []
all_lineups = list(itertools.permutations(batters, 9))

for lineup in all_lineups:
    expected_score = coef["const"]
    for i, batter in enumerate(lineup):
        #expected_score += coef["BB%"][i] * batter["OBP"]
        #expected_score += coef["K%"][i] * batter["K%"]
        expected_score += coef["上壘率"][i] * batter["OBP"]
        expected_score += coef["OPS"][i] * batter["OPS"]
        expected_score += coef["TB/PA"][i] * batter["SLG"]
        expected_score += coef["長打/PA"][i] * (batter["SLG"])
        expected_score += coef["安打/PA"][i] * (batter["AVG"])
    
    lineup_scores.append((expected_score, lineup))
```

```python
#最好打線
# 按照預期得分排序
lineup_scores.sort(reverse=True, key=lambda x: x[0])

# 顯示每條打線和其預期得分
for score, lineup in lineup_scores[:30]:
    print(f"預期得分: {score:.4f} -> " + " -> ".join([b["name"] for b in lineup]))
```

```python
#最差打線
# 按照預期得分排序
lineup_scores.sort(reverse=False, key=lambda x: x[0])

# 顯示每條打線和其預期得分
for score, lineup in lineup_scores[:25]:
    print(f"預期得分: {score:.4f} -> " + " -> ".join([b["name"] for b in lineup]))
```
